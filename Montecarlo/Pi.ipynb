```python
import numpy as np
from numpy.random import *
import math
import matplotlib.pyplot as plt
```

```python
samples = 2000
X = np.random.uniform(-1, 1, samples)
Y = np.random.uniform(-1, 1, samples)
```

```python
accepted = (X ** 2 + Y ** 2) <= 1
rejected = ~accepted
print(np.sum(accepted) * 4 / samples)
```

```python
plt.scatter(X[accepted], Y[accepted], color='blue', label='accepted')
plt.scatter(X[rejected], Y[rejected], color='red', label='rejected')
plt.legend()
```
