```python
import numpy as np
from numpy.random import *
import math
import matplotlib.pyplot as plt
```

```python
samples = 2000
X = np.random.uniform(0, math.pi / 2, samples) # HACK
Y = np.random.uniform(0, 1, samples)
```

```python
accepted = Y <= np.sin(X)
rejected = ~accepted
print(samples * 2 / np.sum(accepted))
```

```python
plt.scatter(X[accepted], Y[accepted], color='blue', label='accepted')
plt.scatter(X[rejected], Y[rejected], color='red', label='rejected')
plt.legend()
```
